```python
%matplotlib inline

import matplotlib.pyplot as plt

# Make a single feature and the output
x = list(range(-100, 101))
y = [num**2 for num in x]
plt.plot(x, y)
plt.title("Relation between x and y")
plt.ylabel("y")
plt.xlabel("x")
plt.show();
```

```python
mse_at_each_split = []
total_points = len(y)

for split_point in range(total_points):
    # Get the left squared errors
    y_left = y[:split_point]
    n_y_left = len(y_left) or 1
    mean_y_left = sum(y_left)/n_y_left
    weight_left = n_y_left/total_points
    left_squared_errors = [(y_val - mean_y_left)**2 for y_val in y_left]
    
    # Get the right squared errors
    y_right = y[split_point:]
    n_y_right = len(y_right) or 1
    mean_y_right = sum(y_right)/n_y_right
    weight_right = n_y_right/total_points
    right_squared_errors = [(y_val - mean_y_right)**2 for y_val in y_right]
    
    # Get the total squared errors
    mse = sum(left_squared_errors + right_squared_errors)
    mse_at_each_split.append(mse)
    
plt.plot(x, mse_at_each_split)
plt.title("MSE for each split point of x")
plt.ylabel("MSE")
plt.xlabel("x");
```

From the chart above, there appear to be two local minima.
